```python
import pandas as pd
import cv2
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import time
```

```python
df = pd.read_csv("../open-image-mapping-resources/source-info/filtered_dataset_IO_local.csv")

print(df.shape)
print(df.columns)
```

```python
image_heights = []
image_widths   = []

start_time = time.perf_counter()

# Image lecture
for i in range(len(df)):

    image_path = df.iloc[i, 6]
    
    # Get images dimensions
    image = cv2.imread(image_path)
    
    image_heights.append(int(image.shape[0]))
    image_widths.append(int(image.shape[1]))

end_time = time.perf_counter()

elapsed_time = end_time - start_time
print(f"Execution time: {elapsed_time:.6f} seconds")
```

Mostrar distribución de tamaños de imagenes

```python
# Explore images dimensions distribution
df = pd.DataFrame({
    'height': image_heights,
    'width': image_widths
})

plt.figure(figsize=(8,6))
sns.boxplot(data=df)
sns.pointplot(data=df, estimator='median', linestyle='none', color='black', markers='d')
plt.title('Distribución de Tamaños de Imagen')
plt.title('Image Dimension Distribution')
plt.ylabel('Pixels')
plt.show()

IMAGE_HEIGHTS_MEDIAN = int(np.median(image_heights))
IMAGE_WIDTH_MEDIAN = int(np.median(image_widths))
print("Height Median: ", IMAGE_HEIGHTS_MEDIAN)
print("Width  Median: ", IMAGE_WIDTH_MEDIAN)
```
